```python
# Carga del dataframe con el archivo steam_games.json.gz
df_steam_games = pd.read_json('dataset/steam_games.json.gz',lines=True,compression='gzip')
df_steam_games.head(10)
```

```python
# Se procede a eliminar todas las filas en donde todos los valores sean NaN y observamos los resultados
df_steam_games = df_steam_games.dropna(how='all')
df_steam_games.head()
```

```python
# Consultamos las columnas
df_steam_games.columns
```

```python
# Eliminamos las columnas que no se requieren
df_steam_games.drop(columns=[ 'publisher', 'genres', 'app_name', 'title', 'url', 'release_date', 'tags', 'reviews_url', 'specs', 'early_access', 'developer'], inplace=True)
df_steam_games.head()
```

```python
valores_unicos_lista = df_steam_games['id'].unique().tolist()

# Imprimir los valores √∫nicos como lista
valores_unicos_lista
```

```python
# reemplazar estos valores cadena 'Starting at $499.00', 'Starting at $449.00' cambiarlos a su precio numerico minimo
df_steam_games['price'] = df_steam_games['price'].apply(lambda x: 499.00 if isinstance(x, str) and '$499.00' in x.lower() else x)
df_steam_games['price'] = df_steam_games['price'].apply(lambda x: 449.00 if isinstance(x, str) and '$449.00' in x.lower() else x)
df_steam_games.head()
```

```python
# Cambiar los valores que no son float por 0.0 en la columna price
df_steam_games['price2'] = pd.to_numeric(df_steam_games['price'], errors='coerce').fillna(0.0)

```

```python
# Filtramos el usuario
user_id='js41637'

 # Filtrar el DataFrame para obtener solo las filas correspondientes al usuario dado
user_data = df_merged_definitivo[df_merged_definitivo['user_id'] == user_id]

# Calcular la cantidad de dinero gastado por el usuario
dinero_gastado = user_data['price2'].sum()

# Calcular el porcentaje de recomendaci√≥n en base a reviews.recommend
total_recomendaciones = user_data['recommend'].count()
recomendaciones_positivas = user_data['recommend'].sum()
porcentaje_recomendacion = (recomendaciones_positivas / total_recomendaciones) * 100 if total_recomendaciones > 0 else 0

# Calcular la cantidad de items
cantidad_items = user_data['item_id'].nunique()

# Crear el diccionario de retorno
user_info = {
    "Usuario": user_id,
    "Dinero gastado": f"{dinero_gastado} USD",
    "% de recomendaci√≥n": f"{porcentaje_recomendacion:.2f}%",
    "Cantidad de items": cantidad_items
}

print(user_info)
```
