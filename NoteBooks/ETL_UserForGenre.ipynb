```python
""" 
   Se decide transformar la columna release_date con el valor de a√±o, pero hay que tener en cuenta que tiene valores (2066 valores nulos) y el formato 
   actual no es de tipo date, para el caso de valores nulos, se contemplan los criterios como que los nulos tienen una distribuci√≥n aleatorios y 
   no tiene otras  columnas relacionadas; se  opta con actualziar los nulos con la mediana de las fechas. """

# Convertir la columna release_date a datetime
df_items['release_date'] = pd.to_datetime(df_items['release_date'], errors='coerce')

# Calcular la mediana de las fechas
# Convertir las fechas a enteros de 64 bits para calcular la mediana
median_timestamp = df_items['release_date'].dropna().astype('int64').median()
median_date = pd.to_datetime(median_timestamp)

# Rellenar los valores nulos con la mediana de fechas
df_items['release_date'] = df_items['release_date'].fillna(median_date)
```

```python
# Asegurando que la columna release_date est√© con formato date, se procede a lo siguiente:
# Adicionar una columna 'year' que extrae el a√±o de la columna 'release_date'
df_items['year'] = df_items['release_date'].dt.year
```

```python
# Carga del dataframe con el archivo users_items.json.gz
data = []
with gzip.open('dataset/users_items.json.gz', 'rt', encoding='utf-8') as file:
    for line in file:
        try:
            json_data = ast.literal_eval(line)
            data.append(json_data)
        except ValueError as e:
            print(f"Error en la l√≠nea: {line}")
            continue
df_users_items = pd.DataFrame(data)

df_users_items 
```

```python
# implementemos el codigo para la API
genero='Simulation'

 # Filtrar el DataFrame por el g√©nero dado
genre_df = merged_df[merged_df['genres'] == genero]

# Si el DataFrame filtrado est√° vac√≠o, retornar el formato esperado con valores nulos
if genre_df.empty:
    print( {"Usuario con m√°s horas jugadas para G√©nero " + genero: None, "Horas jugadas": []})

# Agrupar por usuario y sumar las horas jugadas
user_hours = genre_df.groupby('id')['early_access'].sum()

# Encontrar el usuario con m√°s horas jugadas
max_user = user_hours.idxmax()

# Agrupar por a√±o y sumar las horas jugadas
year_hours = genre_df.groupby('year')['early_access'].sum().reset_index()

# Formatear el resultado en la estructura solicitada
hours_per_year = [{"A√±o": row['year'], "Horas": row['early_access']} for _, row in year_hours.iterrows()]

result = {
    "Usuario con m√°s horas jugadas para G√©nero " + genero: max_user,
    "Horas jugadas": hours_per_year
}

print(result)

```
